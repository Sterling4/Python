```python
x = Iris_set["Sepal.Length"]
y = Iris_set["Petal.Length"]
plt.scatter(x,y)

plt.title("Sepal vs Petal")
plt.axhline()

```

```python
Iris_test = Iris_set[["Petal.Length","Petal.Width","Sepal.Length","Sepal.Width"]]
Iris_grp = Iris_test.groupby(["Petal.Length","Petal.Width","Sepal.Length","Sepal.Width"], as_index=False).mean()
Iris_grp
```

```python
Iris_pivot = Iris_set.pivot(index="Species", columns="Petal.Length")
```
